```python
import pandas as pd
import matplotlib.pyplot as plt

%pylab inline
```

```python
df = pd.read_table('./transcoding_mesurment.tsv')
```

```python
df.describe().T
```

```python
def plot_attrs(x, y):
    fig = plt.figure(figsize=(13, 25))
    for i in range(10):
        plt.subplot(5, 2, i + 1)
        plt.title(u'Вариант {}'.format(i+1))
        plt.xlabel(x)
        plt.ylabel(y)
        start = 67700 + 100 * i
        end = 67699 + 100 * (i + 1)
        plt.scatter(df.loc[start:end, x], df.loc[start:end, y])
    plt.show()
```

```python
plot_attrs('framerate', 'size')
```

```python
plot_attrs('frames', 'size')
```

```python
plot_attrs('bitrate', 'size')
```

```python
plot_attrs('duration', 'size')
```

```python
plot_attrs('width', 'size')
```

```python
plot_attrs('height', 'size')
```

```python
attrs = ['id', 'framerate', 'frames', 'bitrate', 'duration', 'width', 'height', 'size']
```

```python
d = []
```

```python
for i in range(10):
    start = 67700 + 100 * i
    end = 67699 + 100 * (i + 1)
    dff = df.loc[start:end, attrs].copy()
    dff.reset_index(inplace=True)
    dff.drop('index', axis=1, inplace=True)
    d.append(dff)
```

```python
d[0]
```

```python
for i in range(10):
    d[i].to_csv('youtube_var_{}.csv'.format(str(i + 1)))
```
